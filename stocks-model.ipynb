```julia
using CSV, PyPlot, Distributions
```

data found at "https://finance.yahoo.com/quote/NVDA/history?period1=1488964671&period2=1520500671&interval=1d&filter=history&frequency=1d"

```julia
apple=CSV.read("data/apple.csv", nullable=false)
amzn=CSV.read("data/AMZN.csv", nullable=false)
fb=CSV.read("data/FB.csv", nullable=false)
google=CSV.read("data/GOOG.csv", nullable=false)
nfx=CSV.read("data/NFLX.csv", nullable=false)
nvidia=CSV.read("data/NVDA.csv", nullable=false);
#bit=CSV.read("BTC-USD.csv")
```

```julia
subplot(2,1,1)
plot(apple[:, 1], apple[:, 5], label="Apple")
plot(apple[:, 1], fb[:, 5], label="Facebook")
plot(apple[:, 1], nfx[:, 5], label="Netflix")
plot(apple[:, 1], nvidia[:, 5], label="Nvidia")
legend()
title("Adj. Closing [USD] over 1 year")
ylabel("Adjusted Closing Price")
subplot(2,1,2)
plot(apple[:, 1], google[:, 5], label="Google", "b")
plot(apple[:, 1], amzn[:, 5], label="Amazon", "k")
#plot(collect(1:252), amzn[:, 5], label="bitcoin")
legend()
ylabel("Adjusted Closing Price")
xlabel("time [1y]")
```

```julia
#subplot(2,1,1)
plot(apple[:, 1], 100*(apple[:, 5]/apple[1,5]-1), label="apple")
plot(apple[:, 1], 100*(fb[:, 5]/fb[1,5]-1), label="facebook")
plot(apple[:, 1], 100*(nfx[:, 5]/nfx[1,5]-1), label="netflix")
plot(apple[:, 1], 100*(nvidia[:, 5]/nvidia[1,5]-1), label="nvidia")
plot(apple[:, 1], 100*(google[:, 5]/google[1,5]-1), label="google")
plot(apple[:, 1], 100*(amzn[:, 5]/amzn[1,5]-1), label="amazon")
legend()
title("Adjusted Closing Price Increase over 1 year")
ylabel("Adj. Close [%]")
```

```julia
Z=rand(Normal(.4,2.6), 252); #apple: rand(Normal(.2,2), 252);
Xi=0
sig=0.
X=[Xi; [Xi+ (sum(Z[1:i-1])) + sig*Z[i] for i=2:252]];
```

```julia
Z2=rand(Normal(.1, 1.23), 252);
X2=[Xi; [Xi+ (sum(Z2[1:i-1])) + sig*Z2[i] for i=2:252]];
```

```julia
plot(apple[:, 1], X, label="BM model", "k--")
plot(apple[:, 1], X2, label="BM model", "k--")
plot(apple[:, 1], 100*(nvidia[:, 5]/nvidia[1,5]-1), label="NVIDIA", "b")
plot(apple[:, 1], 100*(apple[:, 5]/apple[1,5]-1), label="Apple", "g")
legend()
title("Adjusted Closing Price Increase over 1 year")
ylabel("Adj. Close [%]")
```
